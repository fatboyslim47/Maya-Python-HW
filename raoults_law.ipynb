```python
import numpy as np
import matplotlib.pyplot as plt
# need to solve for the Pi_sat and Pj_sat
#constants are from NIST
T = 25 # degrees celsius
Ap, Bp, Cp = 6.85221, 1064.63, 233.01
Ah, Bh, Ch = 6.87601, 1171.17, 224.41

#Antoine equation for vapor pressure (mmHg)
#will be constant Psat at constant T
def vapor_pressure (A, B, C, T) :
    return 10**(A - (B/(C+T)))
P_pentane_sat = vapor_pressure (Ap, Bp, Cp, T) #compound i
P_hexane_sat = vapor_pressure(Ah, Bh, Ch, T) #compound j
#mole fraction of Pentane range for liquid phase (xE)
xi = np.linspace(0, 1, 100)

#Bubble line
P_bubble_line = xi*P_pentane_sat + (1-xi)*P_hexane_sat
#Dew Line 
yi = xi*P_pentane_sat/P_bubble_line
P_dew_line = 1/((yi/P_pentane_sat) + ((1-yi)/P_hexane_sat))
#Plotting
plt.figure(figsize=(10, 6))
plt.plot(xi, P_bubble_line, label="Bubble Point", color='blue') 
plt.plot(xi, P_dew_line, label="Dew Point", color='red')
plt.title("Bubble and Dew Lines for Pentane and Hexane Mixture at 298K") 
plt.xlabel("Mole Fraction of Pentane") 
plt.ylabel("Pressure (mmHg)")
plt.legend()
plt.grid(True)
plt.show()
print(P_pentane_sat)
print(P_hexane_sat)
```
